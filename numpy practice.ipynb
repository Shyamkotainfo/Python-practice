NumPy arrays have these key properties:


1.   ndim – the number of dimensions (axes) in the array.
2.   shape – the size of the array for each dimension.
3. size – the total number of elements in the array.
4.  dtype – the data type of the elements in the array.

```python
print(f"ndim: {sales_array.ndim}")
print(f"shape: {sales_array.shape}")
print(f"size: {sales_array.size}")
print(f"dtype: {sales_array.dtype}")
```

# ARRAY CREATION
- As an alternative to converting lists, you can create arrays using functions

1. **ones**: Creates an array of ones of a given size, as float by default.
   
    ***np.ones((rows, cols), dtype)***

2. **zeros**: Creates an array of zeros of a given size, as float by default.

    ***np.zeros((rows, cols), dtype)***

3. **arrange**: Creates an array of integers with given start & stop values, and a step size (only stop is required, and is not inclusive).

    ***np.arange(start, stop, step)***

4. **linspace**: Creates an array of floats with given start & stop values with n elements, separated by a consistent step size (stop is inclusive).
    ***np.linspace(start, stop, n)***

5. **reshape**: Changes an array into the specified dimensions, if compatible.
    ***np.array.reshape(rows, cols)***


```python
np.ones(4,)
```

```python
np.zeros((2, 5), dtype=int)
```

```python
np.arange(10)
```

```python
np.arange(5, 10, 1)
```

```python
np.linspace(10, 100, 50) # n elements i.e, 50
```

```python
np.arange(1, 9, 2).reshape(2,2)
```

# Indexing & slicing
**one-dimensional arrays is the same as base Python**

• array[index] – indexing to access a single element (0-indexed)

• array[start:stop:step size] – slicing to access a series of elements (stop is not inclusive)

**two-dimensional arrays requires an extra index or slice**

• array[row index, column index] – indexing to access a single element (0-indexed)

• array[start:stop:step size, start:stop:step size] – slicing to access a series of elements


```python
import numpy as np

product_2d = np.array([[1, 2, 3, 4, 5, 6], [10, 20, 30, 40, 50, 60]])

product_2d
```
