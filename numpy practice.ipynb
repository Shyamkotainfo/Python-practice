# ARRAY CREATION
- As an alternative to converting lists, you can create arrays using functions

1. **ones**: Creates an array of ones of a given size, as float by default.
   
    ***np.ones((rows, cols), dtype)***

2. **zeros**: Creates an array of zeros of a given size, as float by default.

    ***np.zeros((rows, cols), dtype)***

3. **arrange**: Creates an array of integers with given start & stop values, and a step size (only stop is required, and is not inclusive).

    ***np.arange(start, stop, step)***

4. **linspace**: Creates an array of floats with given start & stop values with n elements, separated by a consistent step size (stop is inclusive).
    ***np.linspace(start, stop, n)***

5. **reshape**: Changes an array into the specified dimensions, if compatible.
    ***np.array.reshape(rows, cols)***


```python
np.ones(4,)
```

```python
np.zeros((2, 5), dtype=int)
```

```python
np.arange(10)
```

```python
np.arange(5, 10, 1)
```

```python
np.linspace(10, 100, 50) # n elements i.e, 50
```

```python
np.arange(1, 9, 2).reshape(2,2)
```

# Indexing & slicing
**one-dimensional arrays is the same as base Python**

• array[index] – indexing to access a single element (0-indexed)

• array[start:stop:step size] – slicing to access a series of elements (stop is not inclusive)

**two-dimensional arrays requires an extra index or slice**

• array[row index, column index] – indexing to access a single element (0-indexed)

• array[start:stop:step size, start:stop:step size] – slicing to access a series of elements


```python
import numpy as np

product_2d = np.array([[1, 2, 3, 4, 5, 6], [10, 20, 30, 40, 50, 60]])

product_2d
```

# FILTERING ARRAYS

You can filter arrays by indexing them with a logical test

• Only the array elements in positions where the logical test returns True are returned

```python
sales_array
```

```python
sales_array != 0 # checks the element is not zero
```

```python
sales_array [sales_array != 0] # returns the value which are TRUE
```

You can filter arrays with multiple logical tests

• Use | for or conditions and & for and conditions


```python
sales_array
```

```python
sales_array[(sales_array == 616) | (sales_array < 100)]
 # this retuns an array with elements equal to 616 or less than 100
```

```python
sales_array[(sales_array > 100) & (sales_array < 500)]
# This returns an array with elements greater than 100 and less than 500
```

```python
mask = (sales_array > 100) & (sales_array < 500) # we can store complex filtering in variable
sales_array[mask]
```

You can filter arrays based on values in other arrays

• Use the Boolean array returned from the other array to index the array you want to filter

# Broadcasting

It lets you perform vectorized operations with arrays of different sizes, where NumPy will expand the smaller array to fit the larger one.

- single values (scalars) can be broadcast into arrays of any dimensions. ie., [1]

```python
import numpy as np

# Create two NumPy arrays
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

# Element-wise addition using vectorization
result = a + b

print(result)  # Output: [ 6  8 10 12]
```
